```python
s2='Meeting minutes and lists of the Jewish population of Mena. Statute of the Labor Agricultural Land Artillery, 30.08.1924-03.11.1924, 45 sheets. (https://e-resource.tsdavo.gov.ua/files/21935/)'
```

```python
number_pattern = re.compile('[0-9]+')
```

```python
url_pattern = re.compile('\(https:[^)]+\)')
```

```python
date_pattern = re.compile('[0-9-]+$')
```

```python
s='Case 1 (https://e-resource.tsdavo.gov.ua/files/21935/) 1924'
```

```python
def parse_col1(s):
    number = None
    url = None
    date = None
    m = re.search(number_pattern, s)
    if m is not None:
        span = m.span()
        number = s[span[0]:span[1]]
        s = s[span[1]:]
    m = re.search(url_pattern, s)
    if m is not None:
        span = m.span()
        url = s[(span[0]+1):(span[1]-1)]
        s = s[span[1]:]
    if m is not None:
        span = m.span()
        date = s[span[0]:span[1]]
        s = s[span[1]:]
    return number, url, date
```

```python
def parse_col2(s):
    desc = None
    url = None
    m = re.search(url_pattern, s)
    if m is not None:
        span = m.span()
        url = s[(span[0]+1):(span[1]-1)]
        desc = s[:span[0]]
    return desc, url
```

```python
n, u, d = parse_col1(s)
print (n, u, d)
```

```python
desc, url = parse_col2(s2)
print(desc, '\n', url)
```

```python
def process_sheet(ws):
    for row in ws.iter_rows(min_row=1, max_col=3, max_row=1000):
        s = row[0].value
        num, url, date = parse_col1(s)
        print(num, url, date)
        row[0].value = num
        #row[0].value = f'=hyperlink("{url}", "{num}")'
        row[0].hyperlink = url
        row[2].value = date
        s = row[1].value
        desc, url = parse_col2(s)
        print(desc, url)
        #row[1].value = f'=hyperlink("{url}", "{desc}")'
        row[1].value = desc
        row[1].hyperlink = url
```

```python
process_sheet(ws)
```
